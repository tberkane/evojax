```python
import numpy as np
import matplotlib.pyplot as plt

# Load the best.npz file
data = np.load("best.npz")

# Print the keys in the npz file
print("Keys in the npz file:", data.files)

# Iterate through all arrays in the file and display basic information
for key in data.files:
    array = data[key]
    print(f"\nArray: {key}")
    print(f"Shape: {array.shape}")
    print(f"Data type: {array.dtype}")
    print(f"First few elements: {array.flatten()[:5]}")

    # If the array is 2D and not too large, visualize it
    if len(array.shape) == 2 and array.shape[0] < 100 and array.shape[1] < 100:
        plt.figure(figsize=(10, 8))
        plt.imshow(array, cmap="viridis")
        plt.colorbar()
        plt.title(f"Visualization of {key}")
        plt.show()

# Close the npz file
data.close()
```
